```python
import numpy as np
import pandas as pd
```

```python
t1=pd.read_csv('3.2.2PandaTitanic.csv')
t1.head()
```

```python
t1.Sex.unique()
```

#### MAP:
- map is a method of pandas.Series
- map each values of series.
- iterates over each element of a series.
- where we have categories variables or features.

```python
t1['Sex_Column'] = t1.Sex.map({'female':0,'male':1})
```

```python
t1.loc[0:4,['Sex','Sex_Column']]
```

#### APPLY:
- pandas.Series.apply or pandas.DataFrame.apply
- a DataFrame or Series method
- applies a function along each element of DataFrame or Series

```python
t1.Fare.head()
```

```python
t1['Fare_new']=t1.Fare.apply(np.around,decimals=2)
```

```python
t1.loc[:,['Fare','Fare_new']].head()
```

```python
def cap(x):
    if(x=='female'):
        return x.replace('female','FEMALE')
    else:
        return x
```

```python
t1['Sex_Cap']=t1.Sex.apply(cap)
```

```python
t1.loc[:,['Sex','Sex_Cap']].head()
```

```python
import numpy as np
import pandas as pd
p1=pd.read_csv('3.2.4PandaMlbsalaries.csv')
p1.head()
```

```python
p1.shape
```

```python
p1.Year.value_counts() # Return unique Year

# Similarly p1.Team.value_counts()
# Similarly p1.Position.value_counts()
```

```python
# Retrive information for year 2010
yr_2010=p1[p1.Year==2010]
```

```python
yr_2010.shape
```

```python
yr_2010.head()
```

```python
yr_2010.sort_values('Player').head()  # Return first 5 Data
```

```python
yr_2010.sort_values('Player').tail() # Return last 5 Data
```

```python
yr_2010.sort_index(axis=1,inplace=True)
yr_2010.head()
```

```python
# Find top 5 Highest paid Salaries and corresponding player name
high_salary=yr_2010.sort_values('Salary',ascending=False).head()
```

```python
high_salary[['Player','Salary']].head()
```

```python
# Or
yr_2010.loc[yr_2010.sort_values('Salary',ascending=False).index,['Player','Salary']].head()
```

```python
# Find Highest paid Salaries and corresponding player name for each year
# groupby()
group_yr=p1.groupby('Year')
```

```python
type(group_yr)
```

```python
# agg() to get index of highest salaries of each year
group_yr.Salary.agg(['max'])
```

```python
group_yr.Salary.agg(['max','min'])
```

```python
# Or using apply() to get index of highest salaries of each year
def check(x):
    return x.idxmax()   # argmax() print index of maximum value
group_yr.Salary.apply(check)
```

```python
# Or Using empty list
yr=[]
for i in group_yr:
    yr.append(i[1].Salary.idxmax())
p1.loc[yr]
```

```python
# Or using empty DataFrame
new=pd.DataFrame()
for i in group_yr:
    new=new.append(i[1].sort_values('Salary',ascending=False).head(1))
new
```
